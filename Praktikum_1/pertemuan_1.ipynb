```python
import matplotlib.pyplot as plt
```

```python
candy_names = ['Kit Kat', 'Snickers', 'Milky Way', 'Toblerone', 'Twix']
candy_counts = [52, 39, 78, 13, 78]
explode = [0, 0.1, 0, 0, 0]

plt.figure(figsize=(8, 8))
plt.pie(
    candy_counts,
    labels=candy_names,
    autopct='%1.1f%%',
    startangle=140,
    explode=explode,
    shadow=True
)
plt.title('Presentase Jenis Permen dalam Sekantong Permen', pad=30)
plt.axis('equal')
plt.show()

total_candy = sum(candy_counts)
snickers_count = candy_counts[candy_names.index('Snickers')]
snickers_probability = (snickers_count / total_candy) * 100

print(f'Kemungkinan memilih Snickers secara acak: {snickers_probability:.2f}%')
```
